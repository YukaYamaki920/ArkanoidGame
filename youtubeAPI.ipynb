```python
from googleapiclient.discovery import build

api_key = "AIzaSyAZFbvq_zBoL1rLwi506LQ-urFlPcrfcyk"
youtube = build('youtube', 'v3', developerKey=api_key)
search_query = "ファミレスを享受せよ AND 実況"
next_page_token = None
total_results = []

while True:
    request = youtube.search().list(q=search_query, part="snippet", maxResults=20, pageToken=next_page_token)
    response = request.execute()

    # 結果をtotal_resultsに追加
    total_results.extend(response['items'])

    # 次のページのトークンを取得
    next_page_token = response.get('nextPageToken')

    if not next_page_token:
        break

# total_resultsにすべての検索結果が累積されています
# ここからデータを処理できます
for video in total_results:
    video_id = video['id']['videoId']
    title = video['snippet']['title']
    channel_name = video['snippet']['channelTitle']
    upload_date = video['snippet']['publishedAt']
    channel_id = video['snippet']['channelId']
    channel_request = youtube.channels().list(part="statistics", id=channel_id)
    channel_response = channel_request.execute()
    number_of_subscribers = channel_response['items'][0]['statistics']['subscriberCount']
    print("Title:", title)
    print("Channel:", channel_name)
    print("Date:", upload_date)
    print("URL:", f"https://www.youtube.com/watch?v={video_id}")
    print("Subscribers:", number_of_subscribers)
    print("-" * 40)  # 区切り線
```
